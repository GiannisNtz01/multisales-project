## Read Bronze File

```python
from pyspark.sql.functions import col, trim


df = spark.read \
    .option("header", True) \
    .option("delimiter", "\t") \
    .csv("/Volumes/sales_lakehouse/bronze/bronze_volume/Salesperson.csv")
```

## Rename columns to snake_case

```python

df = df.withColumnRenamed("EmployeeKey","employee_key") \
       .withColumnRenamed("EmployeeID","employee_id") \
       .withColumnRenamed("Salesperson","salesperson_name") \
       .withColumnRenamed("Title","title") \
       .withColumnRenamed("UPN","upn")
```

##Cast numeric columns

```python
df = df.withColumn("employee_key", col("employee_key").cast("int")) \
       .withColumn("employee_id", col("employee_id").cast("int"))
```

## Trim String Columns

```python

for c, t in df.dtypes:
    if t == "string":
        df = df.withColumn(c, trim(col(c)))
```

```python
df.write \
.mode("overwrite") \
.format("delta") \
.saveAsTable("sales_lakehouse.silver.salesperson")

display(df)
```
