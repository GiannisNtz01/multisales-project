## Read Bronze File

```python
df = spark.read \
.option("header", True) \
.option("delimiter", "\t") \
.csv("/Volumes/sales_lakehouse/bronze/bronze_volume/Reseller.csv")
```

## Rename Columns

```python
df = df.withColumnRenamed("ResellerKey","reseller_key") \
       .withColumnRenamed("Business Type","business_type") \
       .withColumnRenamed("Reseller","reseller_name") \
        .withColumnRenamed("City","city") \
       .withColumnRenamed("State-Province","state_province") \
       .withColumnRenamed("Country-Region","country_region")
```

## Cast Types

```python
from pyspark.sql.functions import col

df = df.withColumn("reseller_key", col("reseller_key").cast("int"))
```

## Fix inconsistent values

```python
from pyspark.sql.functions import when

df = df.withColumn(
    "country_region",
    when(col("country_region") == "UK","United Kingdom")
    .otherwise(col("country_region"))
)
```

## Delta Table

```python
df.write \
.mode("overwrite") \
.format("delta") \
.saveAsTable("sales_lakehouse.silver.reseller")

display(df)
```
