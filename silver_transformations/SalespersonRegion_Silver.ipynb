```python
from pyspark.sql import functions as F

bronze_path = "/Volumes/sales_lakehouse/bronze/bronze_volume/SalespersonRegion.csv"

df = (spark.read
      .option("header", True)
      .option("delimiter", "\t")
      .csv(bronze_path))
```

## Rename to snake_case

```python
df = (df.withColumnRenamed("EmployeeKey", "employee_key")
        .withColumnRenamed("SalesTerritoryKey", "sales_territory_key"))
```

```python
(df.write
 .mode("overwrite")
 .option("overwriteSchema", "true")
 .format("delta")
 .saveAsTable("sales_lakehouse.silver.salesperson_region"))
display(spark.table("sales_lakehouse.silver.salesperson_region").limit(20))
```
