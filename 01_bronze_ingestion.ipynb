```python
#comment
from datetime import date
import re

base_volume = "/Volumes/lakehouse/adventureworks_multisales/raw_zone"

raw_root = base_volume                    # our csv files
bronze_root = f"{base_volume}/bronze"     # εδώ θα μπουν οργανωμένα

ingest_date = date.today().isoformat()  #hello world

dbutils.fs.mkdirs(bronze_root)

print("RAW:", raw_root)
print("BRONZE:", bronze_root)
print("INGEST DATE:", ingest_date)
```

```python
display(dbutils.fs.ls(raw_root))
```

```python
raw_files = [f.path for f in dbutils.fs.ls(raw_root) if f.path.lower().endswith(".csv")]

def dataset_name(path: str) -> str:
    base = path.split("/")[-1]
    name = re.sub(r"\.csv$", "", base, flags=re.IGNORECASE)
    return name.lower()

for src in raw_files:
    dataset = dataset_name(src)
    target_dir = f"{bronze_root}/{dataset}/ingest_date={ingest_date}"
    dbutils.fs.mkdirs(target_dir)

    dst = f"{target_dir}/{dataset}.csv"
    dbutils.fs.cp(src, dst)
    print("BRONZE ->", dst)
```
