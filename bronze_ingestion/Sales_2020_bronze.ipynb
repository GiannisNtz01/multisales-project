## **Sales_2020 overview** ##

```python
path = "/Volumes/lakehouse/adventureworks_multisales/raw_zone/Sales_2020.csv"

sales_2020_uncleaned = (spark.read
      .format("csv")
      .option("header", "true")
      .option("sep", "\t")          
      .option("inferSchema", "true")
      .option("mode", "PERMISSIVE")
      .load(path))

display(sales_2020_uncleaned)
```

```python
sales_2020_uncleaned.printSchema()
```

```python
row_count = sales_2020_uncleaned.count()
col_count = len(sales_2020_uncleaned.columns)
row_count, col_count
```

**Check nulls** per column

```python
from pyspark.sql.functions import col, sum as _sum, when

nulls = sales_2020_uncleaned.select([
    _sum(when(col(c).isNull(), 1).otherwise(0)).alias(c)
    for c in sales_2020_uncleaned.columns
])
display(nulls)
```

Check **Empty** Values

```python
from pyspark.sql.functions import trim

empty = sales_2020_uncleaned.select([
    _sum(when(trim(col(c)) == "", 1).otherwise(0)).alias(c)
    for c, t in sales_2020_uncleaned.dtypes if t == "string"
])
display(empty)
```

Check for **duplicates**

```python
total = sales_2020_uncleaned.count()
distinct = sales_2020_uncleaned.distinct().count()
duplicates_fullrow = total - distinct
print(f"duplicate rows: {duplicates_fullrow}")
```
