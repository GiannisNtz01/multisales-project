```python
path = "/Volumes/lakehouse/adventureworks_multisales/raw_zone/Targets.csv"

Targetsbronze_uncleaned = (spark.read
      .format("csv")
      .option("header", "true")
      .option("sep", "\t")          # TAB delimiter
      .option("inferSchema", "true")
      .option("mode", "PERMISSIVE")
      .load(path))

display(Targetsbronze_uncleaned)
```

```python
Targetsbronze_uncleaned.printSchema()
```

**show nulls**

```python
from pyspark.sql import functions as F

count_nulls = Targetsbronze_uncleaned.select([F.sum(F.col(c).isNull().cast("int")).alias(c + "_nulls")
    for c in Targetsbronze_uncleaned.columns])
display(count_nulls)
```

**show duplicates**

```python
total_rows = Targetsbronze_uncleaned.count()
unique_rows = Targetsbronze_uncleaned.dropDuplicates().count()

duplicate_count = total_rows - unique_rows
print("Total duplicated rows:", duplicate_count)
```
