**Diplay our Region table in a right structure**

```python
path = "/Volumes/lakehouse/adventureworks_multisales/raw_zone/Region.csv"

region_uncleaned = (spark.read
      .format("csv")
      .option("header", "true")
      .option("sep", "\t")          # TAB delimiter
      .option("inferSchema", "true")
      .option("mode", "PERMISSIVE")
      .load(path))

display(region_uncleaned)
```

**Checking the data types**

```python
region_uncleaned.printSchema()
```

**Checking if we have duplicated rows**

```python
from pyspark.sql import functions as F

count_nulls = region_uncleaned.select([F.sum(F.col(c).isNull().cast("int")).alias(c + "_nulls")
    for c in region_uncleaned.columns])
display(count_nulls)
```

```python
total_rows = region_uncleaned.count()
unique_rows = region_uncleaned.dropDuplicates().count()

duplicate_count = total_rows - unique_rows
print("Total duplicated rows:", duplicate_count)


```
